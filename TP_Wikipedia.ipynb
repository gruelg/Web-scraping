### Étape 6 : Trouver comment extraire ces informations à l’aide des méthode find() et findAll() et tester les sur les 5 premiers éléments.

```python
data = []
table_body = tables[5].find('tbody')
#print(table_body.find_all('a',href=True))

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('a')
    test = cols[0].get('href')
    cols = [ele.text.strip() for ele in cols]
    cols.append(test)
    data.append([ele for ele in cols if ele]) 

data.pop(0)
data.pop(1)
for i in data :
    print(i)
```

### Étape 8 : Refactorer le code des question précédente afin d’avoir le DataFrame suivant et enregistrer votre dataFrame en .csv ainsi que .xls

```python
d = { 'clubs': [data[i][0] for i in range(4) ], 'url' : [data[i][3] for i in range(4)]}

df = pd.DataFrame(data=d)



print(df)
df.to_csv(r'dataframe.csv',sep=';')
```
